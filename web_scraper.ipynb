### The entire article pages of each result item

```python
search_url = 'https://www.nature.com/search?q=how%20to%20become%20a%20data%20scientist&subject=mathematics-and-computing&date_range=last_30_days&order=relevance'
p = requests.get(search_url)
search_content = p.content
search_soup = BeautifulSoup(search_content, 'html.parser')
```

```python
result_soup = search_soup.find_all('h3',{'class':"c-card__title"})
for i in result_soup:
    i = str(i)
    result_url1 = i.split('href="')[1]
    result_url2 = result_url1.split('" itemprop="url"')[0]
    final_result_url = f"https://www.nature.com{result_url2}"
    q = requests.get(final_result_url)
    final_result_soup = q.content
    final_result_soup = BeautifulSoup(final_result_soup, 'html.parser')
    print(final_result_soup.prettify ,'\n','\n','\n','\n','\n','----------------------------------------------','\n','\n','\n','\n','\n','\n','\n')
```

# Step-2 and Step-3

### extracting the article titles of 100 pages

```python
all_articles_soup = []

for page_no in range (1,101):
    article_url = f"https://www.nature.com/ncomms/articles?searchType=journalSearch&sort=PubDate&type=article&page={page_no}"
    s = requests.get(article_url)
    article_content = s.content
    article_content = BeautifulSoup(article_content , 'html.parser')
    all_articles = article_content.find_all('h3',{'class':"c-card__title"})

    for i in range (0,len(all_articles)):
        all_articles_soup.append(all_articles[i])
points=0        
for t in all_articles_soup:
    t = str(t)
    points = points + 1
    result_url1 = t.split('href="')[1]
    result_url2 = result_url1.split('" itemprop="url"')[0]
    final_result_url = f"https://www.nature.com{result_url2}"
    u = requests.get(final_result_url)
    separate_article_content = u.content
    separate_article_content = BeautifulSoup(separate_article_content , 'html.parser')
    all_separate_articles = separate_article_content.find('h1',{'class':"c-article-title"}).get_text()




    print(str(points) + '. '+all_separate_articles)
```

### Extracting the articles summary

```python
all_articles_soup = []
points=0

for page_no in range (1,101):
    article_url = f"https://www.nature.com/ncomms/articles?searchType=journalSearch&sort=PubDate&type=article&page={page_no}"
    s = requests.get(article_url)
    article_content = s.content
    article_content = BeautifulSoup(article_content , 'html.parser')
    all_articles = article_content.find_all('div',{'data-test':"article-description"})
    for i in range (0,len(all_articles)):
        points = points+1
        print(str(points)+'. '+all_articles[i].get_text())
```
